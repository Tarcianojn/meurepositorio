```python
from faker import Faker
import pandas as pd

# Criar uma instância do Faker
fake = Faker('pt_BR')

# Gerar uma lista de clientes falsos
clientes = []

for _ in range(100):
    cliente = {
        'nome': fake.name(),
        'endereço': fake.address(),
        'email': fake.email(),
        'telefone': fake.phone_number(),
        'data_nascimento': fake.date_of_birth(),
    }
    clientes.append(cliente)

# Converter a lista para um DataFrame do pandas
df_clientes = pd.DataFrame(clientes)

# Mostrar os primeiros registros
df_clientes.head()
```
